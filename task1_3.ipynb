```python
from thinkdsp import CosSignal, SinSignal

cos_sig = CosSignal(freq=440, amp=1.0, offset=0)
sin_sig = SinSignal(freq=880, amp=0.5, offset=0)
```

```python
cos_sig.plot()
```

```python
sin_sig.plot()
```

```python
mixed = sin_sig + cos_sig
mixed.plot()
```

```python
wave = mixed.make_wave(duration=1, start=0, framerate=11025)
```

```python
wave.make_audio()
```

```python
spectrum = wave.make_spectrum()
spectrum.plot()
```
