```python
from thinkdsp import read_wave

wave = read_wave('guitar.wav')
```

```python
wave.make_audio()
```

```python
start = 0.75
duration = 0.5
segment = wave.segment(start, duration)
segment.plot()
decorate(xlabel='Time (s)')
```

```python
spectrum = segment.make_spectrum()
```

```python
spectrum.plot()
decorate(xlabel='Frequency')
```

```python
spectrum.plot(high=8000)
decorate(xlabel='Frequency')
```

```python
spectrum.low_pass(3000)
```

```python
spectrum.plot(high=8000)
decorate(xlabel=Frequency)
```

```python
spectrum.high_pass(3000)
```

```python
spectrum.plot(high=8000)
decorate(xlabel=Freq)
```

```python
filtered = spectrum.make_wave()
```

```python
filtered.normalize()
```

```python
filtered.apodize()
filtered.plot()
decorate(xlabel = "Time (s)")
```

```python
segment.normalize()
segment.apodize()
segment.plot()
decorate(xlavel='Time')
```

```python
segment.make_audio()
```
